```python
import pandas as pd
import numpy as np
import networkx as nx
import seaborn as sns
import folium
import shapefile as shp
import matplotlib.pyplot as plt
from folium import plugins
```

```python
sf = shp.Reader('../map/nycMap/nycb2010.shp')
```

```python
data = pd.read_csv('../data/data_use_outgoing.txt',sep='|')
```

```python
data['startGeoId'] = data.startGeoId.astype(str)
```

```python
m = folium.Map(location=[40.75, -73.99],zoom_start=12,tiles='Mapbox Bright')
m.choropleth('../map/BoroughMap.geojson',fill_color='none')

points = []
for x,y in data.iterrows():
    points.append([y[1],y[2],y[3]])
    
m.add_children(plugins.HeatMap(points,
                               min_opacity=0.1,
                               max_zoom=30,
                               radius=7.5,
                               blur=1))
```
